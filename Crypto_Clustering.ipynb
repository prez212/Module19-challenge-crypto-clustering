```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90,
    title= 'Crypto Market Data: Returns in %',
    ylabel= 'Return %',
    xlabel= 'Crypto Name',
    value_label= 'Return per period %'
)
```

---

### Prepare the Data
This section prepares data for the k-means algortihim,.
1. Use StandardScaler () docule from sckit-learn to normalize CSV file data
2. Create DF to contain scaled data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaled_crypto = StandardScaler().fit_transform(df_market_data)
scaled_crypto
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(scaled_crypto, columns=df_market_data.columns)

# Copy the crypto names from the original data
df_market_data_scaled["coin_id"] = df_market_data.index

# Set the coin_id column as index
df_market_data_scaled = df_market_data_scaled.set_index("coin_id")

# Display sample data
df_market_data_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
list_k = list(range(1,12))
list_k
```

```python
# Create an empty list to store the inertia values
inertia_values= []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:

for i in list_k:
    model =KMeans(n_clusters=i) # 1. Create a KMeans model using the loop counter for the n_clusters
    model.fit(df_market_data_scaled)# 2. Fit the model to the data using `df_market_data_scaled`
    inertia_values.append(model.inertia_)# 3. Append the model.inertia_ to the inertia list

inertia_values
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_dic = {'k': list_k, 'inertia': inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_dic)
elbow_df
```

```python
# Plot a line chart with all the inertia values computed with the different values of k to visually identify the optimal value for k.

elbow_chart= elbow_df.hvplot.line(
    x= "k",
    y="inertia",
    title="Elbow for K-Means",
    xticks=list_k
)
elbow_chart
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Judging from the chart 4 appears to be the best value for 'k'.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
kmeans_clusters = model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
kmeans_clusters
```

```python
# Create a copy of the DataFrame
df_market_data_scaled_copy = df_market_data_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_data_scaled_copy['Prediction']=kmeans_clusters

# Display sample data
df_market_data_scaled_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
market_scaled_chart= df_market_data_scaled_copy.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by = "Prediction",
    hover_cols = "coin_id"
)
market_scaled_chart
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
PCA = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
clust_pca = PCA.fit_transform(df_market_data_scaled)


# View the first five rows of the DataFrame. 
clust_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.

PCA.explained_variance_ratio_
```

```python
#Calculate the total
sum(PCA.explained_variance_ratio_)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 85.5% 

```python
# Create a new DataFrame with the PCA data.
df_market_data_pca = pd.DataFrame(
    clust_pca,
    columns=["PCA1", "PCA2","PCA3"])

# Copy the crypto names from the original data
df_market_data_pca['coin_id']=df_market_data.reset_index()['coin_id']

# Set the coinid column as index
df_market_data_pca.set_index('coin_id', inplace=True)

# Display sample data
df_market_data_pca.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_pca = list(range(1, 12))
k_pca
```

```python
# Create an empty list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:

for i in k_pca:
    pca_model = KMeans(n_clusters=i) # 1. Create a KMeans model using the loop counter for the n_clusters
    pca_model.fit(df_market_data_pca)# 2. Fit the model to the data using `df_market_data_pca`
    inertia_pca.append(pca_model.inertia_)# 3. Append the model.inertia_ to the inertia list

inertia_pca
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_dic = {'k_pca': k_pca, 'inertia': inertia_pca
                 }
# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(pca_elbow_dic)
df_elbow_pca.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

elbow_pca_chart = df_elbow_pca.hvplot.line(
    x='k_pca',
    y='inertia',
    title = 'PCA Elbow Curve',
    xticks = k_pca
)
elbow_pca_chart
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No. 

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
crypto_pca_model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
crypto_pca_model.fit(df_market_data_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
pca_prediction = crypto_pca_model.predict(df_market_data_pca)
# Print the resulting array of cluster values.
pca_prediction
```

```python
# Create a copy of the DataFrame with the PCA data
df_market_data_pca_prediction = df_market_data_pca.copy()

# Add a new column to the DataFrame with the predicted clusters
df_market_data_pca_prediction['pca_prediction']= pca_prediction

# Display sample data
df_market_data_pca_prediction.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
pca_market_chart = df_market_data_pca_prediction.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="pca_prediction",
    hover_cols=['coin_id']
)
pca_market_chart
```
